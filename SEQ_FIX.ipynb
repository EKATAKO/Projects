```python
def are_files_equal(original_path,converted_path):
    try:
        with open(original_path, 'r') as original_file:
            data_orig = original_file.read()
        
        with open(converted_path, 'r') as converted_file:
            data_conv = converted_file.read()
        
        original_file_name = os.path.basename(original_path)
            
        if data_orig.strip() == data_conv.strip():
            return None
        else:
            return print(original_file_name)

    except FileNotFoundError:
        return False        
```

```python
def SEQ_CONVERSION(file_path):

    # Read the data from the original file
    with open(file_path, 'r') as original_file:
        data = original_file.read()   

    # Split the data into lines
    rows = data.split('\n')

    # Initialize a flag to track whether we should collect lines
    collect_lines = False

    # Initialize a list to store the lines before and after "xxxxx"
    lines_before_xxxxx = []
    lines_after_xxxxx = []

    # To store lines which have been convert to read the correct sort codes
    modified_lines = []


    # Iterate through the lines and collect lines after "xxxxx"
    for row in rows:
        if "XXXXXX" in row:
            collect_lines = True
        elif collect_lines:
            lines_after_xxxxx.append(row)

    # Iterate through the lines and collect lines before "xxxxx"
    for row in rows:
        if "XXXXXX" in row:
            collect_lines = False
            lines_before_xxxxx.append(row)
        elif collect_lines:
            lines_before_xxxxx.append(row)

    # Loop through each line in lines_after_xxxxx
    for line in lines_after_xxxxx:
        # Initialize a variable to keep track of the index of the first alphabet character
        index_of_first_alphabet = None

        # Find the index of the first alphabet character (uppercase) in the word
        for index, char in enumerate(line):
            if char in alphabet:
                index_of_first_alphabet = index
                break

        # Check if there are 33 characters before the first alphabet character
        if index_of_first_alphabet == 33 or index_of_first_alphabet != 27:
            # Iterate through branches
            for branch in branches:
                if branch in line:
                    # Fetch the sort code from lookup_data (replace this with your actual data retrieval)
                    sort_code = lookup_data.loc[lookup_data['Branch Name'] == branch, 'Sort Code'].values[0]
                    # Modify the line
                    modified_line = line[:6] + sort_code + line[12:]
                    modified_lines.append(modified_line)
        elif index_of_first_alphabet == 27:
            # Iterate through branches
            for branch in branches:
                if branch in line:
                    # Fetch the sort code from lookup_data (replace this with your actual data retrieval)
                    sort_code = lookup_data.loc[lookup_data['Branch Name'] == branch, 'Sort Code'].values[0]
                    # Modify the line
                    modified_line = line[:6] + sort_code + line[6:]
                    modified_lines.append(modified_line)

    ## Combine the data splits to become final data
    Final = lines_before_xxxxx + modified_lines
    
    Final_data = '\n'.join(Final)
    
    original_file_name = os.path.basename(file_path)

    
    # Define the output directory (you can change this as needed)
    output_directory = 'C:\\Users\\KATAKOF\\OneDrive - Universal Merchant Bank Limited\\Documents\\Output\\Actual\\'

    # Construct the path for the new file with the original file name
    new_file_path = os.path.join(output_directory, original_file_name)
    
    Converted.append(Final_data)
    
    
    # Export the data to the new file
    with open(new_file_path, 'w') as new_file:
        new_file.write(Final_data)

#     print(f"Data from '{original_file_name}' has been exported to '{new_file_path}'")
```
