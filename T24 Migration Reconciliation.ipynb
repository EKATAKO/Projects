```python
def trim_string_columns(data):
    if isinstance(data, pd.DataFrame):
        string_columns = data.select_dtypes(include='object').columns
        data[string_columns] = data[string_columns].apply(lambda x: x.str.strip())
    return data

# Trim columns first in r_14 dataset
r14_columns_trimmed = trim_string_columns(r14_data)
```

```python
# Trim string columns in r_14 dataset
r22_columns_trimmed = trim_string_columns(r22_data)
```

```python
# Extract common observations based on "@ID" column
common_rows = pd.merge(r14_columns_trimmed, r22_columns_trimmed, on= identifier, how="inner")

# Extract column names that are common between the datasets
common_columns = r14_data.columns.intersection(r22_data.columns)
```

```python
merged_columns = []

for column in columns_to_merge:
    column_x = f"{column}_x"
    column_y = f"{column}_y"

    try:
        merged_column = np.where(
            common_rows[column_x] != common_rows[column_y],
            np.where(
                (common_rows[column_x] != "") & (common_rows[column_y] != ""),
                common_rows[column_x].astype(str) + "|" + common_rows[column_y].astype(str),
                np.where(common_rows[column_x] == "", " |" + common_rows[column_y].astype(str), common_rows[column_x].astype(str))
            ),
            ""
        )

        # Append the merged column to the list
        merged_columns.append(pd.DataFrame(merged_column, columns=[f"{column}_z"]))

    except KeyError:
        continue

# Concatenate all merged columns at once
merged_columns_df = pd.concat([common_rows] + merged_columns, axis=1)
```
